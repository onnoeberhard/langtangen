```python
class Parent:
    def __init__(self):
        print("parent init")
        
        
class Child(Parent):
    def __init__(self):
        super().__init__()
        print("child init")
        
c = Child()
```

```python
isinstance(c, Parent)
```

```python
issubclass(Child, Parent)
```
