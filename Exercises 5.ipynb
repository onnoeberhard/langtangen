# Exercise 5.17: Write table to file

```python
def write_table_to_file(f, xmin, xmax, nx, ymin, ymax, ny, width=10, decimals=None, filename="517.dat"):
    x = linspace(xmin, xmax, nx)
    y = linspace(ymax, ymin, ny)
    z = f(*np.meshgrid(x, y))
    with open("stuff/" + filename, 'w') as f:
        for i, row in enumerate(z):
            f.write(f"{y[i]:13.6g}")
            for e in row:
                f.write(f"{e:13.6g}")
            f.write("\n")
        f.write("%13s" % "")
        for x in x:
            f.write(f"{x:13.6g}")
        f.write("\n")

def test_write_table_to_file():
    write_table_to_file(f=lambda x, y: x + 2*y,
                        xmin=0, xmax=2, nx=4,
                        ymin=-1, ymax=2, ny=3)
    # Load text in file and compare with expected results
    with open("stuff/517.dat") as infile:
        computed = infile.read()
    expected = """\
            2            4      4.66667      5.33333            6
          0.5            1      1.66667      2.33333            3
           -1           -2     -1.33333    -0.666667            0
                         0     0.666667      1.33333            2\n"""
    assert computed == expected
    
test_write_table_to_file()
```

# Exercise 5.23: Vectorize the Midpoint rule for integration

```python
import builtins

def midpointint1(f, a, b, n=1000):
    h = (b - a) / n
    r = 0
    for i in range(1, n+1):
        r += h * f(a - h/2 + i*h)
    return r

def midpointint2(f, a, b, n=1000):
    h = (b - a) / n
    return h * builtins.sum(f(a - h/2 + i*h) for i in range(1, n+1))

def midpointint3(f, a, b, n=1000):
    h = (b - a) / n
    i = arange(1, n+1)
    return h * np.sum(f(a - h/2 + i*h))
```

# Exercise 5.25: Implement Lagranges interpolation formula

```python
def p_L(x, xp, yp):
    x = np.asarray(x)
    k = arange(len(xp))
    return sum(yp[k] * L_k(x, k, xp, yp), axis=1)

def L_k(x, k, xp, yp):
    i = empty([len(k) - 1, len(k)], dtype=int)
    for k_ in k:
        i[:, k_] = np.delete(k, k_)
    g = x.reshape(-1, 1, 1) - np.resize(xp[i], (x.size, xp[i].shape[0], xp[i].shape[1]))
    return prod(g / (xp[k] - xp[i]), axis=1)
```

```python
xp = array([1, 3, 4])
yp = array([1, 3, 2])
plt.figure()
plt.scatter(xp, yp)
plt.show()
```

```python
p_L(array([1, 3]), xp, yp)
```

```python
def test_p_L():
    xp = array([1, 3, 4])
    yp = array([1, 3, 2])
    error = sum(abs(p_L(xp[i], xp, yp) - yp[i]) for i in range(len(xp)))
    assert error < 1e-12
    
test_p_L()
```

# Exercise 5.37: Animate a sequence of approximations to $\pi$

```python
def pathlength(x, y):
    return sum(sqrt((x[i] - x[i - 1])**2 + (y[i] - y[i - 1])**2) for i in range(1, len(x)))

def pi_approx(N):
    x = [1/2 * cos(2 * pi * i / N) for i in range(N)]
    y = [1/2 * sin(2 * pi * i / N) for i in range(N)]
    path = pathlength(x, y)
    return x, y, path

N = arange(4, 100)
    
import matplotlib as mpl
from matplotlib.animation import FuncAnimation

fig = plt.figure()
ln, = plt.plot([], [])
plt.axis('equal')
plt.axis([-1, 1, -1, 1])

def draw(t):
    x, y, path = pi_approx(t)
    x.append(x[0])
    y.append(y[0])
    plt.title(fr"$\pi \approx {path}$")
    ln.set_data(x, y)
    return ln,

anim = FuncAnimation(fig, draw, N, interval=300, blit=True)
mpl.rc('animation', html='html5')

anim
```

# Exercise 5.39: Animate the evolution of Taylor polynomials

```python
from scipy.special import factorial

def animate_series(fk, M, N, xmin, xmax, ymin, ymax, n, exact):
    x = linspace(xmin, xmax, n)
    t = arange(0, N - M + 1)
    S = lambda t: sum(fk(k, x) for k in range(0, M + t))
    
    fig = plt.figure()
    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot(x, exact(x))
    ln, = plt.plot([], [])
    
    def draw(t):
        plt.title(f"{t + 1} summands")
        ln.set_data(x, S(t))
        return ln,

    return FuncAnimation(fig, draw, t, interval=100, blit=True)
```

# Exercise 5.47: Plot a piecewise constant function

```python
def plot_piecewise(data, xmax):
    x = np.zeros(2*len(data))
    y = np.zeros(2*len(data))
    for i, point in enumerate(data):
        x[2*i - 1] = x[2*i] = point[0]
        y[2*i] = y[2*i + 1] = point[1]
    x[-1] = xmax
    y[-1] = y[-2] = data[-1][1]
    plt.plot(x, y)
    plt.show()

plot_piecewise([(0, -1), (1, 0), (1.5, 2)], 3)
```

# Exercise 5.54: Verify linear algebra results

```python
def test_distributivity():
    n = 4
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    C = np.random.rand(n, n)
    result1 = (A + B) @ C
    result2 = A @ C + B @ C
    assert abs(result1 - result2).max() < 1e-6

test_distributivity()
```

```python
def test_associativity():
    n = 4
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    C = np.random.rand(n, n)
    result1 = (A @ B) @ C
    result2 = A @ (B @ C)
    assert abs(result1 - result2).max() < 1e-6
    
test_associativity()
```

```python
def test_rank():
    n = 4
    A = np.random.rand(n, n)
    result1 = np.linalg.matrix_rank(A)
    result2 = np.linalg.matrix_rank(A.T)
    assert abs(result1 - result2) < 1e-6
    
test_rank()
```

```python
def test_det():
    n = 4
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    result1 = np.linalg.det(A @ B)
    result2 = np.linalg.det(A) * np.linalg.det(B)
    assert abs(result1 - result2) < 1e-6
    
test_det()
```

```python
def test_eigenvalues():
    n = 4
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    result1 = np.linalg.eigvals(A)
    result2 = np.linalg.eigvals(A.T)
    assert abs(result1 - result2).max() < 1e-6
    
test_eigenvalues()
```
